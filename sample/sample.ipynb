```python
import numpy as np


kernel = np.array([[-1, 0, 1], [0, 0, 0], [1, 0, -1]])
```

```python
def convolve2d(image, kernel):
    kernel = np.flipud(np.fliplr(kernel))
    output = np.zeros_like(image)
    n = kernel.shape[0]
    m = kernel.shape[1]
    image_padded = np.zeros((image.shape[0] + n-1, image.shape[1] + m-1))
    image_padded[n//2:(n//2)*-1, m//2:(m//2)*-1] = image
    for x in range(image.shape[1]):
        for y in range(image.shape[0]):
            output[y, x]=(kernel * image_padded[y: y+n, x: x+n]).sum()

    return output
x = convolve2d(sample, kernel)
```

```python
plot.imshow(x, cmap='gray')
```

```python

x = convolve2d(kernel=kernel, image=x)
```

```python
plot.imshow(x, cmap='gray')
```
